# Step 3: Data Cleaning and Tags Creations


```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from tqdm import tqdm

nlp = spacy.load("en_core_web_sm", disable=["parser", "ner", "tagger"])
def clean_and_extract_tags_pipe(texts):
    docs = nlp.pipe((str(text).lower() for text in texts if pd.notna(text)), batch_size=1000)
    return [
        ', '.join([token.text for token in doc if token.text.isalnum() and token.text not in STOP_WORDS]) 
        for doc in docs]
    
columns_to_extract_tags_from = ['Product_Name', 'Category_Name']

for column in columns_to_extract_tags_from:
    try:
        print(f"Processing column: {column}")
        train_data[f'{column}_tags'] = [
            ', '.join([token.text for token in doc if token.text.isalnum() and token.text not in STOP_WORDS]) 
            for doc in tqdm(
                nlp.pipe((str(text).lower() for text in train_data[column] if pd.notna(text)), batch_size=1000),
                total=len(train_data[column]))]
    except Exception as e:
        print(f"Error processing column {column}: {e}")

train_data['Tags'] = train_data[[f'{col}_tags' for col in columns_to_extract_tags_from]].fillna('').apply(
    lambda row: ', '.join(row), axis=1)

train_data['Tags'] = train_data[columns_to_extract_tags_from].apply(lambda row: ', '.join(row), axis=1)
# Specify the file name
output_file = "Final_Product_Dataset_with_Tags.csv"
train_data.to_csv(output_file, index=False)
print(f"Processed data saved to '{output_file}'.")
```

# Content Base Recommendation system 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from scipy.sparse import vstack
import faiss
from tqdm import tqdm

train_data['content'] = train_data['Tags'].fillna('')  # Handle NaN values

batch_size = 10000 
max_features = 100000 
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=max_features)
tfidf_vectorizer.fit(train_data['content'])
tfidf_matrix_batches = []

for i in tqdm(range(0, train_data.shape[0], batch_size), desc="TF-IDF Batches"):
    batch = train_data['content'][i:i+batch_size]
    tfidf_batch = tfidf_vectorizer.transform(batch)  # Only transform, do not fit again
    tfidf_matrix_batches.append(tfidf_batch)

tfidf_matrix_content = vstack(tfidf_matrix_batches)
n_components = 100  # Further reduce dimensions for memory efficiency
svd = TruncatedSVD(n_components=n_components)
tfidf_reduced_chunks = []
for i in tqdm(range(0, tfidf_matrix_content.shape[0], batch_size), desc="SVD Batches"):
    chunk = tfidf_matrix_content[i:i+batch_size]
    reduced_chunk = svd.fit_transform(chunk)  # Perform SVD on the chunk
    tfidf_reduced_chunks.append(reduced_chunk)
tfidf_reduced = np.vstack(tfidf_reduced_chunks).astype(np.float32)
faiss.normalize_L2(tfidf_reduced)  # Normalize the data for cosine similarity

index = faiss.IndexFlatIP(n_components)  # Dimension matches n_components

# Add vectors to the FAISS index using tqdm to monitor progress
for i in tqdm(range(0, tfidf_reduced.shape[0], batch_size), desc="FAISS Indexing"):
    chunk = tfidf_reduced[i:i+batch_size]  # Process in chunks
    index.add(chunk)  # Add chunk to the FAISS index

query_vector = tfidf_reduced[0].reshape(1, -1) 
k = 10 
distances, indices = index.search(query_vector, k)

print("Top similar items indices:", indices)
print("Similarity scores:", distances)
```

# Function To Recommend Products for Content Base

```python
def get_content_based_recommendations(product_index, faiss_index, product_embeddings, train_data, top_n=10):
    """
    Returns the top N most similar products based on FAISS similarity search.

    :param product_index: Index of the query product
    :param faiss_index: FAISS index
    :param product_embeddings: Array of product embeddings
    :param train_data: Original product dataset
    :param top_n: Number of recommendations to return
    :return: DataFrame with recommended products
    """
    query_vector = product_embeddings[product_index].reshape(1, -1)  # Reshape for FAISS
    distances, indices = faiss_index.search(query_vector, top_n + 1)  # +1 to exclude itself

    # Exclude the query product itself
    similar_product_indices = indices[0][1:]  # Skip the first one as it's the query itself

    return train_data.iloc[similar_product_indices]

# Example: Get recommendations for the first product
recommended_products = get_content_based_recommendations(0, index, tfidf_reduced, train_data)
recommended_products
```

# Collaborative Filtering (User Item Similarity)

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

user_search_history = {
    "NehaTembhe": ["laptop", "gaming"],
    "Sameer": ["smartphone", "camera"],
    "Riya": ["headphones", "wireless"],
    "Diksha": ["laptop", "office"],
    "Vaishnavi": ["tablet", "e-reader"],
}
product_data = train_data[["Product_ID", "Product_Name", "Category_Name", "Selling_Price"]].copy()
product_data.loc[:, "Product_Description"] = product_data["Product_Name"] + " " + product_data["Category_Name"]
tfidf_vectorizer = TfidfVectorizer(stop_words="english")
product_tfidf_matrix = tfidf_vectorizer.fit_transform(product_data["Product_Description"])

def recommend_products_cf(user_id, top_n=5):
    if user_id not in user_search_history:
        print("No search history for this user.")
        return None

    search_terms = " ".join(user_search_history[user_id])
    user_query_tfidf = tfidf_vectorizer.transform([search_terms])
    similarity_scores = cosine_similarity(user_query_tfidf, product_tfidf_matrix).flatten()
    top_product_indices = similarity_scores.argsort()[-top_n:][::-1]

    # Return the most relevant products
    return product_data.iloc[top_product_indices]

# Example: Get recommendations for 'NehaTembhe'
print("Recommended products for NehaTembhe:")
recommend_products_cf("NehaTembhe")
```

# Hybrid Recommendations


```python
def hybrid_recommendations(user_id, top_n=5):
    """
    Combines content-based filtering (FAISS) and collaborative filtering to generate hybrid recommendations.
    """
    # Step 1: Content-Based Recommendations
    query_vector = tfidf_reduced[0].reshape(1, -1)  # Example query vector
    distances, indices = index.search(query_vector, top_n * 2)
    content_recommendations = product_data.iloc[indices[0]].copy()
    content_recommendations["Content_Score"] = distances[0]

    # Step 2: Collaborative Filtering Recommendations
    collaborative_recommendations = recommend_products_cf(user_id, top_n * 2)
    if collaborative_recommendations is not None and not collaborative_recommendations.empty:
        collaborative_recommendations = collaborative_recommendations.copy()
        collaborative_recommendations["Collaborative_Score"] = np.linspace(1, 0, len(collaborative_recommendations))
    else:
        collaborative_recommendations = product_data.sample(n=top_n * 2, random_state=42).copy()
        collaborative_recommendations["Collaborative_Score"] = 0  # Fallback score

    # Step 3: Combine Recommendations
    combined_recommendations = pd.concat([content_recommendations, collaborative_recommendations]).drop_duplicates()

    # Step 4: Normalize Scores
    combined_recommendations["Content_Score"] = combined_recommendations["Content_Score"] / combined_recommendations["Content_Score"].max()
    combined_recommendations["Collaborative_Score"] = combined_recommendations["Collaborative_Score"].fillna(0)
    combined_recommendations["Collaborative_Score"] = combined_recommendations["Collaborative_Score"] / combined_recommendations["Collaborative_Score"].max()

    # Step 5: Calculate Hybrid Score
    combined_recommendations["Hybrid_Score"] = (
        combined_recommendations["Content_Score"] * 0.7 +  # Higher weight for content-based
        combined_recommendations["Collaborative_Score"] * 0.3
    )

    # Step 6: Sort and Return Top N
    combined_recommendations = combined_recommendations.sort_values(by="Hybrid_Score", ascending=False)
    return combined_recommendations.head(top_n)

# Example usage
print("Hybrid Recommended Products for NehaTembhe:")
recommended_products = hybrid_recommendations("NehaTembhe", top_n=10)
recommended_products
```

# Finding Accuracy of Algorithms 

### Evaluation and Visualization of Recommendation Algorithm Performance


### Evaluation Metrics: RMSE, MAE, Precision, Recall, F1-Score, Accuracy


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, precision_score, recall_score, f1_score, accuracy_score

def evaluate_recommendation(actual_ratings, predicted_ratings, algorithm_name, threshold=0.5):
    """
    Evaluate a recommendation algorithm using MAE, RMSE, Precision, Recall, F1-Score, Accuracy,
    and visualize Actual vs. Predicted Ratings.

    Parameters:
    - actual_ratings: List of actual ratings (ground truth)
    - predicted_ratings: List of predicted ratings (numerical scores from the algorithm)
    - algorithm_name: String representing the algorithm name (e.g., 'Content-Based')
    - threshold: Cutoff value for relevance classification (default: 0.5 for binary labels)

    Returns:
    - A dictionary of all metrics.
    """
    # Calculate MAE and RMSE
    mae = mean_absolute_error(actual_ratings, predicted_ratings)
    rmse = np.sqrt(mean_squared_error(actual_ratings, predicted_ratings))

    # Convert predicted ratings to binary labels (e.g., relevant or irrelevant)
    predicted_labels = [1 if score >= threshold else 0 for score in predicted_ratings]
    actual_labels = [1 if rating >= threshold else 0 for rating in actual_ratings]

    # Calculate Precision, Recall, F1-Score, and Accuracy
    precision = precision_score(actual_labels, predicted_labels, zero_division=0)
    recall = recall_score(actual_labels, predicted_labels, zero_division=0)
    f1 = f1_score(actual_labels, predicted_labels, zero_division=0)
    accuracy = accuracy_score(actual_labels, predicted_labels)

    # Print Metrics
    print(f"\nMetrics for {algorithm_name}:")
    print(f"MAE: {mae}")
    print(f"RMSE: {rmse}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1-Score: {f1}")
    print(f"Accuracy: {accuracy}")

    # Plot Actual vs. Predicted Ratings
    plt.figure(figsize=(6, 4))
    plt.scatter(actual_ratings, predicted_ratings, alpha=0.5, color='blue')
    x_min, x_max = min(actual_ratings), max(actual_ratings)
    plt.plot([x_min, x_max], [x_min, x_max], 'r--', label='Perfect Prediction')  # Line y=x

    # Dotted lines connecting actual to predicted values
    for i in range(len(actual_ratings)):
        plt.plot([actual_ratings[i], actual_ratings[i]], [predicted_ratings[i], actual_ratings[i]], 
                 color='gray', linestyle='dotted')

    # Labels and title
    plt.xlabel("Actual Ratings")
    plt.ylabel("Predicted Ratings")
    plt.title(f"{algorithm_name}: Actual vs. Predicted Ratings")
    plt.text(max(actual_ratings), min(predicted_ratings), 
             f'MAE: {mae:.3f}\nRMSE: {rmse:.3f}', 
             fontsize=12, verticalalignment='bottom', horizontalalignment='right', color='green')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Return metrics as a dictionary for later use
    return {
        "MAE": mae,
        "RMSE": rmse,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1,
        "Accuracy": accuracy
    }
```
